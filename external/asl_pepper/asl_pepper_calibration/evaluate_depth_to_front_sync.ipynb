```python
import numpy as np
from pyniel.ros_tools.quickros import *
```

```python
a = QuickMultiSubscriber([
    {'topic': "/pepper_robot/camera/front/image_raw",
     'msg_class': sensor_msgs.msg.Image},
    {'topic': "/pepper_robot/camera/depth/image_raw",
     'msg_class': sensor_msgs.msg.Image}
    ], n_messages=10000)
```

```python
front_stamps = [msg.header.stamp for msg in a.messages[0]]
depth_stamps = [msg.header.stamp for msg in a.messages[1]]
```

```python
min_delta_ms = [np.min([abs((a - b).to_nsec() * 0.000001) for b in depth_stamps])
                for a in front_stamps]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
h = plt.hist(min_delta_ms[1:-1],bins=100)
plt.title("Distribution of $\Delta$t between depth and front cameras (n=10000)")
plt.xlabel("$\Delta$t [ms]")
plt.ylabel("N measured")
#plt.vlines([np.mean(min_delta_ms[1:-1])], 0, np.max(h[0]), color='r')
```

```python
import os
save_dir = "./evaluate_depth_to_front_sync"
save_path = os.path.join(save_dir, "measured_min_delta_ms.txt")
```

```python
if not os.path.exists(save_dir):
       print("{} doesn't exist. trying to make dir".format(save_dir))
       os.makedirs(save_dir)
np.savetxt(save_path,min_delta_ms)
print("{} saved.".format(save_path))
```
